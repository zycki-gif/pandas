```python
import pandas as pd
```

```python
df=pd.read_csv('https://pycourse.s3.amazonaws.com/temperature.csv')
```

```python
df['date']=pd.to_datetime(df['date'])
df
```

```python
df=df.set_index('date')
```

```python
#plot de linhas
df.plot();
```

```python
#plot de linhas:tamanho
df.plot(figsize=(10,5));
```

```python
#plot de linhas:grid
df.plot(figsize=(10,5),grid=True)
```

```python
#plot de linhas:style
df.plot(style='-o',linewidth=1.5,figsize=(10,5),grid=True)
#df.plot(style='--',linewidth=2,figsize=(10,5),grid=True)
#df.plot(style='-.',linewidth=2,figsize=(10,5),grid=True)
```

```python
#plot de linhas:cores
df.plot(style='-o',linewidth=1.5,color='#20cb09',figsize=(10,5),grid=True)
```

```python
#plot de barras
df['classification'].value_counts().plot.bar(figsize=(10,5),rot=0)
dg=df['classification'].value_counts()
```

```python
#plot de barras
dg.plot(kind='bar',figsize=(10,5),rot=30);
```

```python
#plot de pizza
dg.plot(kind='pie',autopct='%1.1f%%',shadow=True,figsize=(10,5));
```
